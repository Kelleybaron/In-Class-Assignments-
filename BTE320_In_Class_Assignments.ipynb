```python
class Person:
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  def display(self, formal = True):
    """
    Displays the person name.
    If formal = True, shows 'Last name, First name'
    If formal = False, shows 'First name, Last name'
    """
    if formal:
      print(f"{self.last_name}, {self.first_name}")
    else:
      print(f"{self.first_name}, {self.last_name}")

  def __str__(self):
    return f"{self.first_name} {self.last_name}"
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class Greeting"

name = Greeting("Jane", "Doe")
print(name)
name.display()
name.display(formal=True)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  @property
  def name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  @name.setter
  def name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} has already enrolled in {newCourse}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self):

    self.p1 = Point(0,0)
    self.p2 = Point(5,0)
    self.p3 = Point(0,5)
    self.p4 = Point(5,5)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height


z = Rectangle()
print(f"The area of the first rectangle is: {z.area()}")

p1_test = Point(0,0)
p2_test = Point(5,0)
p3_test = Point(0,5)
p4_test = Point(5,5)


z2 = Rectangle()
print(f"The area of the second rectangle is: {z2.area()}")
```

```python
from warnings import WarningMessage
class Employee:
  def __init__(self, name, wage, hours_worked):
    self.name = name
    self.__wage = wage
    self.hours_worked = hours_worked

  def get_wage(self):
    return self.__wage

  def set_wage(self, wage):
    if wage < 0:
      print("Negative wages are not allowed")
    else:
      self.__wage = wage

  def calculate_pay(self):
    return self.__wage * self.hours_worked

  def display_info(self):
    print(f"Employee Name: {self.name}")
    print(f"Hourly Wage: {self.__wage}")
    print(f"Hours Worked: {self.hours_worked}")
    print(f"Weekly Pay: ${self.calculate_pay()}")

class Manager(Employee):
  def __init__(self, name, wage, hours_worked, bonus):
    super().__init__(name, wage, hours_worked)
    self.bonus  = bonus

  def calculate_pay(self):
    return super().calculate_pay() + self.bonus

  def display_info(self):
    print(f"Manager Name: {self.name}")
    print(f"Hourly Wage: {self.get_wage()}") # Changed to use get_wage()
    print(f"Hours Worked: {self.hours_worked}")
    print(f"Bonus: ${self.bonus}")
    print(f"Weekly Pay: ${self.calculate_pay()}")

employee1 = Employee("John", 20, 40)
manager1 = Manager("Sarah", 30, 40, 500)

employee1.display_info()
print()

manager1.display_info()
```
